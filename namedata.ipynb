In this notebook, I go through the CSV file of names given in the USA year 2023 to find out how many children have been given the name Armi and how many times versions of the name Armi appear in the dataset.

```python
import requests
import pandas as pd
```

```python
# here is my data
usanames_df = pd.read_csv("yob2023.csv")
usanames_df
```

```python
# cheking my datatypes
usanames_df.dtypes
```

```python
# adding headers to dataframe

usanames_df.columns = ['name', 'sex', 'givento']
usanames_df.head()
```

```python
# want to find armi-name within all the names. str-contains doesn't work
# usanames_df['name'].str.contains("armi")

# chatGPT told me to create new variable and it worked!

armi_name = usanames_df[usanames_df['name'].str.contains("armi", case=False, na=False)]
armi_name
```

```python
# also sorting this

armi_name = armi_name.sort_values(by='givento', ascending=True)
armi_name
```

```python
# making a csv to download

csv_file_path = 'armi_name.csv'
armi_name.to_csv(csv_file_path, index=False)
from IPython.display import FileLink
FileLink(csv_file_path)
```
